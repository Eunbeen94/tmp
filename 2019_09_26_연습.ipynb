# Facets   : 어떠한 기준으로 창을 나누고자 할 때 사용한다.

# Theme 

그림을 보다 더 이쁘게 만들고 싶을 때 사용하여 그림을 이쁘게 만들 수 있다. 

```r
ggplot(data = mpg) + 
  geom_point(mapping = aes(x = displ, y = hwy)) +
  facet_grid(drv ~ cyl) +theme_bw()+
theme(axis.line=element_line(size=0.8,color="black"),
     panel.grid.minor=element_blank())
```

```r
ggplot(data = mpg) + 
  geom_point(mapping = aes(x = displ, y = hwy)) +
  facet_grid(drv ~ cyl) +theme_bw()+
theme(axis.line=element_line(size=0.8,color="black"),
     panel.grid.minor=element_blank(),
     panel.border=element_blank(),
     text=element_text(size=25))
```

# Improve readability using pipe operators

흐름의 순서대로 함수를 사용하는 것이 가능
예를들어 
f(s(d())) => d에 적용되고 s에 적용한다음에 f에 적용 이라는 의미인데 이를 d()  s()  f()  로 표현이 가능


```r
by_dest <- group_by(flights, dest)
delay <- summarise(by_dest,
  count = n(),
  dist = mean(distance, na.rm = TRUE),
  delay = mean(arr_delay, na.rm = TRUE)
)
delay <- filter(delay, count > 20, dest != "HNL")
```

```r
delay[1:10,]
```

```r
flights %>% 
  group_by(year, month, day) %>% 
  summarise(mean = mean(dep_delay))
```

```r
flights %>% 
  group_by(year, month, day) %>% 
  summarise(count=n(),
           dist=mean(distance, na.rm=TRUE))%>%
ggplot(aes(x=dist,y=count))+
geom_bar(stat="identity")
```
